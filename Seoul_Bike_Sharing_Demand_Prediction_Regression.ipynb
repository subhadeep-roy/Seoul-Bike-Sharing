## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

* relationship between visibility and rented bike count.
* relationship between windspeed and rented bike count.
* relationship between rainfall and rented bike count.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

* Null hypothesis: There is no linear relationship between Rented Bike Count and Visibility.
* Alternative hypothesis: There is a linear relationship between Rented Bike Count and Visibility.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Perform hypothesis testing
from scipy import stats
# Null hypothesis: There is no linear relationship between Rented Bike Count and Visibility.
# Alternative hypothesis: There is a linear relationship between Rented Bike Count and Visibility.
slope, intercept, r_value, p_value, std_err = stats.linregress(dataset['Visibility (10m)'], dataset['Rented Bike Count'])

# Print the results
print('P-value:', p_value)

# Interpret the results
if p_value < 0.05:
  print('We can reject the null hypothesis.')
else:
  print('We cannot reject the null hypothesis.')
````

##### Which statistical test have you done to obtain P-Value?

* Here we used simple linear regression hypothesis test to obtain p-value.

##### Why did you choose the specific statistical test?

 Simple linear regression is used to assess the relationship between two variables: 'Visibility (10m)' and 'Rented Bike Count'. The choice of simple linear regression is based on the assumption that the relationship between these two variables can be adequately represented by a straight line.

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments

import pandas as pd
from scipy.stats import zscore


# Convert numeric columns to numeric type, coercing errors to NaN for non-numeric values
numeric_df = dataset.apply(pd.to_numeric, errors='coerce')

# Calculate Z-scores for each column
z_scores = zscore(numeric_df)

# Define a threshold for identifying outliers (e.g., Z-score > 3 or Z-score < -3)
threshold = 3

# Find outliers in the dataset
outliers = (abs(z_scores) > threshold).any(axis=1)

# Remove outliers from the dataset
dataset_1 = dataset[~outliers]

# Display the dataset without outliers
print(dataset_1)
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ML Model - 1 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

# Create the Random Forest Regressor
rf_regressor = RandomForestRegressor()
# Define the hyperparameter distributions
param_dist = {
    'n_estimators': randint(50, 200),
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': randint(2, 10),
    'min_samples_leaf': randint(1, 5),
    'max_features': ['auto', 'sqrt', 'log2']
}

# Use Randomized Search for hyperparameter tuning
random_search = RandomizedSearchCV(rf_regressor, param_distributions=param_dist, n_iter=10, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
random_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = random_search.best_params_

# Train the model with the best hyperparameters on the full training set
best_model = RandomForestRegressor(**best_params)
best_model.fit(X_train, y_train)
# Fit the Algorithm

# Predict on the model
y_predict = best_model.predict(X_test)

y_pred = np.round(sc_y.inverse_transform((y_predict).reshape(-1,1)))

y_test_new = sc_y.inverse_transform(y_test)

pd.concat([pd.DataFrame(y_test_new), pd.DataFrame(y_pred)], axis=1)
````
